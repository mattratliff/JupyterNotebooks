```python
from matplotlib import pyplot as plt
```

<h3>Line Chart</h3>

```python
years=[1930, 1940, 1950, 1960, 1970, 1980, 1990]
```

```python
gdp=[22.3, 435.3, 644.6, 3323.5, 8020.4, 14223.3, 24223.5]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.show()
```

<h3>Bar Chart</h3>

```python
cities=["miami", "new york", "orlando", "las vegas", "austin"]
```

```python
population=[323323, 5235234, 2324633, 223323, 66493]
```

```python
plt.bar(range(len(cities)), population)
plt.title("Population")
plt.ylabel("Population")
plt.xticks(range(len(cities)), population)
plt.show()
```

<h3>Scatterplot</h3>

```python
#Scatterplots are good for visualizing the relationships betwee two paired sets of data
```

```python
timespentinstore=[12, 33, 10, 35, 18, 45, 32, 44, 15]
```

```python
numberofitemsbought=[5, 32, 8, 25, 16, 22, 46, 29, 4]
```

```python
labels=['1', '2', '3', '4', '5', '5', '6', '7', '8', '9']
```

```python
plt.scatter(numberofitemsbought, timespentinstore)
plt.title("Time in store vs. Number items purchased")
plt.xlabel("# of items brought")
plt.ylabel("amount of time in store")
plt.axis("equal")
plt.show()
```

```python
#you can see a general relationship that exists (the more time spent in the store the more items purchased)
```
