<b>Create a function to retrieve only the data for the letter needed</b>

```python
def getdata(letter, dataset):
    datatoreturn = []
    for lettervalues in dataset:
        if lettervalues[1] == letter:
            datatoreturn.append(lettervalues)
    return datatoreturn
```

<b>Create a function to combine all observations for a given letter (class)</b>

```python
def getpixeldata(letter, dataset):
    datatoreturn = []
    for lettervalues in dataset:
        if lettervalues[1] == letter:
            datatoreturn.append(lettervalues[6:134])
    return datatoreturn
```

```python
def buildaverages(letters, dataset):
    container = {}
    for letter in letters:
        container.update({letter: np.mean(getpixeldata(letter, dataset), axis=0)})
    return container
```

```python
def classifyletter(lettersample, averages):
    letter = ''
    minvalue = 1000
    for i, averageletter in enumerate(averages):
        minavgvalue = getvalue(lettersample, averageletter)
        print(letters[i], minavgvalue)
        if minavgvalue < minvalue:
            minvalue = minavgvalue
            letter = letters[i]
    return letter
```

```python
def getvalue(lettersample, model):
    totalsum = 0
    for i, pixel in enumerate(model):
        if lettersample[i] == 1:
            pixeldiff = np.absolute(np.subtract(lettersample[i], model[i]))
            totalsum = totalsum + pixeldiff
    return totalsum
```
